# Preprocessing

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

### Cleaning Data

```python
df = pd.read_csv('../Data Science Projects/cosmetic.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.loc[pd.notnull(df['ingredients'])]
df.info()
```

```python
df.Label[df['Label'] == 'moisturizing-cream-oils-mists'] = str('Moisturizer')
df.Label[df['Label'] == 'cleanser'] = str('Cleanser')
df.Label[df['Label'] == 'facial-treatments'] = str('Treatment')
df.Label[df['Label'] == 'face-mask'] = str('Face Mask')
df.Label[df['Label'] == 'eye-treatment-dark-circle-treatment'] = str('Eye Cream')
df.Label[df['Label'] == 'sunscreen-sun-protection'] = str('Sunscreen')
```

```python
df.Label.unique()
```

```python
df.head()
```

```python
df_2 = df['name'].drop_duplicates()
df = df.loc[df_2.index, :].reset_index()
```

```python
df.info()
```

```python
df.loc[df['rank'].isnull()]
```

```python
df['rank'].fillna(0, inplace = True)
df.info()
```

```python
df.head()
```

```python
df.drop(['URL', 'index'], axis = 1, inplace = True)
```

```python
pattern = re.compile(r"(\d+).\d+")
for i in range(len(df)):
    df['price'][i] = re.findall(pattern, df['price'][i])[0]

df['price'] = pd.to_numeric(df['price'])
```

```python
df.head()
```

```python
# skin_type
pattern = re.compile(r"([a-zA-Z]+)\\n")
for i in range(len(df)):
    df['skin_type'][i] = re.findall(pattern, df['skin_type'][i])
```

```python
df.head()
```

```python
# list column dummies
df_2 = df['skin_type'].str.join('|').str.get_dummies()
df_3 = df.join(df_2).drop('skin_type', axis = 1)
df = df_3
```

```python
df.head()
```

```python
## tokenize ingredients
a = [t.split('\r\n\r\n') for t in df['ingredients']]
pattern = ['\r\n', '-\w+: ', 'Please', 'No Info', 'This product', 'Visit']
```

```python
for i in range(len(df)):
    Num = len(a[i])
    for j in range(Num):
        if all(x not in a[i][j] for x in pattern):
            df['ingredients'][i] = a[i][j]
```
